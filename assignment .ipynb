# (5.a) Remove all duplicates observations

```python
Covid.duplicated()
```

```python
Covid=Covid.drop_duplicates()
```

# (5.b) Find missing values in all columns

```python
Covid.isna()
```

```python
Covid.isna().sum()
```

# (5.c) Remove all observations where continent column value is missing

```python
Covid["continent"].isna()
```

```python
Covid=Covid.dropna(subset=["continent"])
```

```python
Covid["continent"].isna()
```

```python
Covid["continent"].isna().sum()
```

# (5.d) Fill all missing values with 0

```python
Covid=Covid.fillna(0)
Covid
```

# (6.a) Convert date column in datetime format using pandas.to_datetime

```python
Covid["date"]=pd.to_datetime(Covid["date"])
```

```python
Covid["date"]
```

# (6.b) Create new column month after extracting month data from date column.

```python
Covid["month"]=pd.DatetimeIndex(Covid["date"]).month
```

```python
Covid["month"]
```

# (7.a) Find max value in all columns using groupby function on 'continent' column

```python
Covid.groupby("continent").max().reset_index()
```

# (7.b) Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

```python
df_groupby=Covid
```

```python
df_groupby
```

# (8.a) Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby["total_deaths_to_total_cases"]=df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby
```

# (9.a) Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
sns.distplot(df_groupby["gdp_per_capita"], kde=False, color="red", bins=30)
```

```python
sns.scatterplot(x="gdp_per_capita",
               y="total_cases",
               data=df_groupby)
```

# (9.c) Plot Pairplot on df_groupby dataset.
